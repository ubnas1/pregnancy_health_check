```python
import pandas as pd
from pathlib import Path
import geopandas
```

```python
# Readdata
file_path = Path("Resources/unused/maternal.csv")
df = pd.read_csv(file_path, encoding='latin-1')

# Display sample data
df.head(10)
```

```python
geo_df = geopandas.read_file("Resources/unused/countries.geojson")
```

```python
joined_gdf = geo_df.merge(df, on="Country")
joined_gdf.to_file('Resources/unused/geojson.geojson', driver='GeoJSON')
```
